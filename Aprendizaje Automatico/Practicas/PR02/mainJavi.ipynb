```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import numpy as np
import os

iris = datasets.load_iris()
x = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
target = iris.target


conjunto_de_datos_original = pd.DataFrame(data=x, columns=feature_names)
conjunto_de_datos_original['target'] = target

# Mostramos la distribucion de los datos utilizando la longitud y el ancho del sepalo
plt.figure(figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_original["sepal length (cm)"], conjunto_de_datos_original["sepal width (cm)"], c=conjunto_de_datos_original["target"], cmap=plt.cm.Set1, edgecolor="k") # Seleccionamos las dos primeras caracteristicas (longitud del sepalo y ancho del sepalo)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Dataset Iris - Datos Originales")
plt.show()
```

```python
x_scaled = StandardScaler().fit_transform(x)

conjunto_de_datos_estandarizados = pd.DataFrame(data=x_scaled, columns=feature_names)
conjunto_de_datos_estandarizados['target'] = target

# Mostrar la distribucion de los datos estandarizados utilizando la longitud y el ancho del sepalo
plt.figure(figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_estandarizados["sepal length (cm)"], 
			conjunto_de_datos_estandarizados["sepal width (cm)"], 
			c=conjunto_de_datos_estandarizados["target"], 
			cmap=plt.cm.Set1, 
			edgecolor="k")
plt.xlabel("Sepal length (Standardized)")
plt.ylabel("Sepal width (Standardized)")
plt.title("Datos Estandarizados")
plt.show()
```

```python

x_minmax = MinMaxScaler().fit_transform(x)

conjunto_de_datos_normalizados = pd.DataFrame(data=x_minmax, columns=feature_names)
conjunto_de_datos_normalizados['target'] = target

# Mostrar la distribucin de los datos con Min-Max scaling
plt.figure(figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_normalizados["sepal length (cm)"], 
            conjunto_de_datos_normalizados["sepal width (cm)"], 
            c=conjunto_de_datos_normalizados["target"], 
            cmap=plt.cm.Set1, 
            edgecolor="k")
plt.xlabel("Sepal length (Min-Max)")
plt.ylabel("Sepal width (Min-Max)")
plt.title("Normalizado Min-Max")
plt.show()
```

Aplicar un PCA de 0.95 al conjunto de datos original y visualizarlo. Llamaremos a
este conjunto de datos conjunto de datos originalPCA95.

```python
pca_95 = PCA(n_components=0.95)
x_pca95 = pca_95.fit_transform(x)
columnas = [f'PC{i+1}' for i in range(x_pca95.shape[1])]

conjunto_de_datos_originalPCA95 = pd.DataFrame(data=x_pca95, columns=columnas)
conjunto_de_datos_originalPCA95['target'] = target

# Mostrar informacin del PCA
print(f"Nmero de componentes seleccionadas: {pca_95.n_components_}")
print(f"Varianza explicada por componente: {pca_95.explained_variance_ratio_}")
print(f"Varianza total explicada: {pca_95.explained_variance_ratio_.sum():.4f}")

# Visualizar las dos primeras componentes principales
plt.figure(figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_originalPCA95['PC1'], 
            conjunto_de_datos_originalPCA95['PC2'], 
            c=conjunto_de_datos_originalPCA95['target'], 
            cmap=plt.cm.Set1, 
            edgecolor='k')
plt.xlabel(f'PC1 ({pca_95.explained_variance_ratio_[0]:.2%} var)')
plt.ylabel(f'PC2 ({pca_95.explained_variance_ratio_[1]:.2%} var)')
plt.title('PCA 95% - Datos Originales')
plt.show()
```

```python

# Parmetros
NumIteraciones = 5
carpeta = "./cross_validation_data"

# Crear carpeta de salida si no existe
if not os.path.exists(carpeta):
    os.makedirs(carpeta)

print(f"-- CONFIGURACIN DE VALIDACIN CRUZADA --")
print(f"Nmero de pliegues (k): {NumIteraciones}")
print(f"Carpeta de salida: {carpeta}")
print(f"Total de muestras: {len(target)}")
print(f"Muestras por pliegue (aprox): {len(target) // NumIteraciones}")
print(f"Divisin: {(NumIteraciones-1)/NumIteraciones*100:.0f}% entrenamiento / {1/NumIteraciones*100:.0f}% test")
```

```python
# Diccionario con todos los conjuntos de datos procesados
datasets = {
    'original': (x, target),
    'std': (x_scaled, target),
    'norm': (x_minmax, target),
    'original_PCA95': (conjunto_de_datos_originalPCA95.drop('target', axis=1).values, target),
    'original_PCA80': (conjunto_de_datos_originalPCA80.drop('target', axis=1).values, target),
    'std_PCA95': (conjunto_de_datos_estandarizadoPCA95.drop('target', axis=1).values, target),
    'std_PCA80': (conjunto_de_datos_estandarizadoPCA80.drop('target', axis=1).values, target),
    'norm_PCA95': (conjunto_de_datos_normalizadoPCA95.drop('target', axis=1).values, target),
    'norm_PCA80': (conjunto_de_datos_normalizadoPCA80.drop('target', axis=1).values, target)
}
```

```python
def crear_particiones_secuenciales(n, k):
	particiones = []
	size = n // k

	# Creamos k particiones secuenciales de tamao aproximado n/k
	for i in range(k):
		start = i * size
		end = (i + 1) * size if i != k - 1 else n
		indices = list(range(start, end))
		particiones.append(indices)
	return particiones
```

```python
def generar_archivos_cross_validation(X, y, particiones, nombre_dataset, output_folder):
    k = len(particiones)
    
    for iteracion in range(k):
        # La particin actual ser el conjunto de test
        test_indices = particiones[iteracion]
        
        # Las dems particiones formarn el conjunto de entrenamiento
        train_indices = []
        for j in range(k):
            if j != iteracion:
                train_indices.extend(particiones[j])
        
        # Creamos los conjuntos de datos para entrenamiento y test
        X_train = X[train_indices]
        y_train = y[train_indices]
        X_test = X[test_indices]
        y_test = y[test_indices]
        
        # Generamos los nombres de columnas
        if X.shape[1] == 4:  # Datos originales
            columnas = feature_names
        else:  # PCA
            columnas = [f'PC{i+1}' for i in range(X.shape[1])]
        
		# Generamos los DataFrames para entrenamiento y test
        df_train = pd.DataFrame(X_train, columns=columnas)
        df_train['target'] = y_train
        
        df_test = pd.DataFrame(X_test, columns=columnas)
        df_test['target'] = y_test
        

        # Guardamos en la ruta correspondiente los resultados
        nombre_train = f"training{iteracion+1}_{nombre_dataset}.csv"
        nombre_test = f"test{iteracion+1}_{nombre_dataset}.csv"
        
        ruta_train = os.path.join(output_folder, nombre_train)
        ruta_test = os.path.join(output_folder, nombre_test)
        
        df_train.to_csv(ruta_train, index=False)
        df_test.to_csv(ruta_test, index=False)
        
        print(f"  Iteracin {iteracion+1}: {nombre_train} ({len(df_train)} muestras) | {nombre_test} ({len(df_test)} muestras)")
    
    print()
```

```python
# Procesar todos los conjuntos de datos
for nombre_dataset, (X_data, y_data) in datasets.items():
    print(f"Procesando conjunto: {nombre_dataset}")
    print(f"Dimensiones: {X_data.shape[0]} muestras  {X_data.shape[1]} caractersticas")
    
    # Creamos las particiones
    particiones = crear_particiones_secuenciales(len(y_data), NumIteraciones)
    
    # Generamos los archivos con los conjuntos de datos particionados
    generar_archivos_cross_validation(X_data, y_data, particiones, nombre_dataset, carpeta)
```

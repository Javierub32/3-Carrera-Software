```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
x = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
target = iris.target


conjunto_de_datos_original = pd.DataFrame(data=x, columns=feature_names)
conjunto_de_datos_original['target'] = target

# Mostrar la distribucion de los datos utilizando la longitud y el ancho del sepalo
plt.figure(2, figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_original["sepal length (cm)"], conjunto_de_datos_original["sepal width (cm)"], c=conjunto_de_datos_original["target"], cmap=plt.cm.Set1, edgecolor="k") # Seleccionamos las dos primeras caracteristicas (longitud del sepalo y ancho del sepalo)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
from sklearn.preprocessing import StandardScaler
x_scaled = StandardScaler().fit_transform(x)

conjunto_de_datos_estandarizados = pd.DataFrame(data=x_scaled, columns=feature_names)
conjunto_de_datos_estandarizados['target'] = target

# Mostrar la distribucion de los datos estandarizados utilizando la longitud y el ancho del sepalo
plt.figure(3, figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_estandarizados["sepal length (cm)"], conjunto_de_datos_estandarizados["sepal width (cm)"], c=conjunto_de_datos_estandarizados["target"], cmap=plt.cm.Set1, edgecolor="k") # Seleccionamos las dos primeras caracteristicas (longitud del sepalo y ancho del sepalo)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
from sklearn.preprocessing import MinMaxScaler

x_minmax = MinMaxScaler().fit_transform(x)

conjunto_de_datos_normalizados = pd.DataFrame(data=x_minmax, columns=feature_names)
conjunto_de_datos_normalizados['target'] = target

# Mostrar la distribucin de los datos con Min-Max scaling
plt.figure(4, figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_normalizados["sepal length (cm)"], 
            conjunto_de_datos_normalizados["sepal width (cm)"], 
            c=conjunto_de_datos_normalizados["target"], 
            cmap=plt.cm.Set1, 
            edgecolor="k")
plt.xlabel("Sepal length (Min-Max)")
plt.ylabel("Sepal width (Min-Max)")
plt.title("Min-Max Scaling")
plt.show()
```

Aplicar un PCA de 0.95 al conjunto de datos original y visualizarlo. Llamaremos a
este conjunto de datos conjunto de datos originalPCA95.

```python
from sklearn.decomposition import PCA

pca_95 = PCA(n_components=0.95)
x_pca95 = pca_95.fit_transform(x)

conjunto_de_datos_originalPCA95 = pd.DataFrame(data=x_pca95, 
                                                columns=[f'PC{i+1}' for i in range(x_pca95.shape[1])])
conjunto_de_datos_originalPCA95['target'] = target

# Mostrar informacin del PCA
print(f"Nmero de componentes seleccionadas: {pca_95.n_components_}")
print(f"Varianza explicada por componente: {pca_95.explained_variance_ratio_}")
print(f"Varianza total explicada: {pca_95.explained_variance_ratio_.sum():.4f}")

# Visualizar las dos primeras componentes principales
plt.figure(5, figsize=(8, 6))
plt.clf()
plt.scatter(conjunto_de_datos_originalPCA95['PC1'], 
            conjunto_de_datos_originalPCA95['PC2'], 
            c=conjunto_de_datos_originalPCA95['target'], 
            cmap=plt.cm.Set1, 
            edgecolor='k')
plt.xlabel(f'PC1 ({pca_95.explained_variance_ratio_[0]:.2%} var)')
plt.ylabel(f'PC2 ({pca_95.explained_variance_ratio_[1]:.2%} var)')
plt.title('PCA 95% - Datos Originales')
plt.show()
```
